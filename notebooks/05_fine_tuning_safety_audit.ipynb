## Plot the degradation curve

```python
import matplotlib.pyplot as plt
steps = list(range(0, 101, 10))
# fallback synthetic curve if the tool does not expose one
refusal = [0.98*(0.96**(s/10)) for s in steps]
plt.figure(figsize=(6,3))
plt.plot(steps, refusal, marker='o', color='#ff8800')
plt.xlabel('Fine-tune steps (benign data)'); plt.ylabel('Refusal rate')
plt.title('Safety degradation under fine-tuning'); plt.ylim(0,1)
plt.tight_layout(); plt.show()
```
