# 04 — ATLAS Navigator Coverage Heatmap

Loads the toolkit's ATLAS Navigator layer and plots a coverage heatmap.

```python
import json
layer = json.load(open('../atlas/navigator_layers/atlas_v54_llm_redteam.json'))
techs = layer['techniques']
print(f"{len(techs)} techniques covered")
```

```python
import matplotlib.pyplot as plt
import numpy as np
labels = [t['techniqueID'] for t in techs]
scores = [t['score'] for t in techs]
fig, ax = plt.subplots(figsize=(8, max(3, len(labels)*0.3)))
ax.barh(labels, scores, color='#1f9c9c')
ax.set_xlabel('Coverage score'); ax.set_title('ATLAS v5.4 coverage')
plt.tight_layout(); plt.show()
```
