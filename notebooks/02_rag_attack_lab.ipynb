````python
collection = _InMemoryCollection()
collection.add(['Doc about data retention: 90 days.', 'Doc about backups.'], ['d1','d2'])
````

````python
exploit = RAGPoisoningExploit(collection=collection, generate=lambda q,docs: ' '.join(docs)[:200])
query = 'what is the data retention policy'
print('Before:', f"{exploit.retrieval_success(query):.0%} poisoned")
````

## Inject phantom document and re-measure

````python
exploit.inject(query)
print('After: ', f"{exploit.retrieval_success(query):.0%} poisoned")
findings = exploit.evaluate(query)
print(f'Findings: {len(findings)}')
````
