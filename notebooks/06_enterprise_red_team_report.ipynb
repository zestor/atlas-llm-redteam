# 06 — Enterprise Red Team Report

End-to-end: run scanners, aggregate findings, generate an executive summary.

```python
import sys, json; sys.path.append('..')
from atlas.techniques.AML_T0051_prompt_injection.exploit import PromptInjectionExploit, _demo_vulnerable_target
```

```python
exploit = PromptInjectionExploit(target=_demo_vulnerable_target)
findings = exploit.run_all()
print(f'Collected {len(findings)} findings')
```

## Aggregate by severity and OWASP category

```python
from collections import Counter
by_sev = Counter(f.severity for f in findings)
by_owasp = Counter(f.owasp_category for f in findings)
print('By severity:', dict(by_sev))
print('By OWASP:', dict(by_owasp))
```

## Generate executive summary markdown

```python
lines = ['# Red Team Executive Summary', '', f'Total findings: {len(findings)}', '']
for sev, n in by_sev.items():
    lines.append(f'- **{sev}**: {n}')
summary = chr(10).join(lines)
open('red_team_summary.md','w').write(summary)
print(summary)
```
