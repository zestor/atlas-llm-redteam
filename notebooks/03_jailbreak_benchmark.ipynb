## Plot ASR per family

```python
import matplotlib.pyplot as plt
fams = list(asr.keys()); vals = [asr[f] for f in fams]
plt.figure(figsize=(6,3))
plt.bar(fams, vals, color='#1f9c9c')
plt.ylabel('ASR'); plt.title('Jailbreak ASR by family'); plt.ylim(0,1)
plt.tight_layout(); plt.show()
```
